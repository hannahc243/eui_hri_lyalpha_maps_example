```python
from sunpy.map import Map
from sunpy.net import Fido, attrs as a
import astropy.units as u
import matplotlib.pyplot as plt
import numpy as np
import sunpy_soar
from astropy.coordinates import SkyCoord
from sunpy.time import parse_time
```

```python
query = Fido.search(a.Time('2024-01-31 02:00', '2024-01-31 03:00'), a.Instrument.eui, a.Level(2), a.soar.Product("EUI-HRILYA1216-IMAGE"))
```

```python
query
```

```python
res = Fido.fetch(query)
```

```python
lyalpha_map_sequence = Map(res, sequence=True)
```

```python
%matplotlib notebook
lyalpha_map_sequence.plot()
```

```python
lyalpha_maps = Map(res)
```

```python
lyalpha_running_diff_maps = Map(
    [m - prev_m.quantity for m, prev_m in zip(lyalpha_map_sequence[1:], lyalpha_map_sequence[:-1])],
    sequence=True
)
```

```python
lyalpha_running_diff_maps.plot()
```

```python
lyalpha_running_diff_maps[40].plot()
```

```python
bl = SkyCoord(-63*u.arcsec, 388*u.arcsec, frame=lyalpha_maps[0].coordinate_frame)
tr = SkyCoord(136*u.arcsec, 525*u.arcsec, frame=lyalpha_maps[0].coordinate_frame)
```

```python
lyalpha_submaps = [lyalpha_maps[i].submap(bottom_left=bl, top_right=tr) for i in range(len(lyalpha_maps))]
```

```python
lyalpha_maps[40].plot()
lyalpha_maps[40].draw_quadrangle(bottom_left=bl, top_right=tr)
```

```python
lyalpha_submaps[40].rotate().plot()
```

```python
lyalpha_ts = [lyalpha_submaps[i].data.sum() for i in range(len(lyalpha_submaps))]
time = [lyalpha_submaps[i].date.datetime for i in range(len(lyalpha_submaps))]
```

```python
fig,ax=plt.subplots()
ax.plot(time, lyalpha_ts)
ax.set_ylabel('Flux (cts/s)')
```

```python
from stixpy.net.client import STIXClient  # Registers the STIXClient with Fido
from stixpy.product import Product
```

```python
query_stx = Fido.search(a.Instrument.stix, a.Time('2024-01-31 02:00', '2024-01-31 03:00'), a.stix.DataType.sci)
```

```python
query_stx
```

```python
res_stx = Fido.fetch(query_stx)
```

```python
stx_spec = Product(res_stx)
```

```python
stx_spec.plot_spectrogram()
```

```python
stx_spec.plot_timeseries(energy_indices = [[1,6], [7,11]])
plt.legend()
```

```python
fig,ax=plt.subplots(2,1, sharex=True)

ax[0].plot(time, lyalpha_ts)
ax[0].set_ylabel('Flux (cts/s)')
ax[0].set_title('EUI/HRI Lyman alpha ')

stx_spec.plot_timeseries(axes=ax[1], energy_indices = [[1,6], [7,11]])
ax[1].set_xlim([parse_time('2024-01-31 02:30').datetime,parse_time('2024-01-31 02:50').datetime])
ax[1].set_yscale('linear')
ax[1].set_ylim([0,0.3])
ax[1].legend()
ax[1].set_title('STIX')
```
